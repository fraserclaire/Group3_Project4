```python
cancer_capita_all_2021 = []
lung_bronchus_capita_2021 = []
for x in range(len(cancer_df)):
    all_capita_2021 = cancer_df['2021'][x]/cancer_df['2021_Population'][x]
    lb_capita_2021 = cancer_df['2021 Lung & Bronchus'][x]/cancer_df['2021_Population'][x]
    
    cancer_capita_all_2021.append(all_capita_2021)
    lung_bronchus_capita_2021.append(lb_capita_2021)
```

```python
cancer_df['2021_AllSites_perCapita'] = cancer_capita_all_2021
cancer_df['2021_LungBronchus_perCapita'] = lung_bronchus_capita_2021
cancer_df.head()
```

```python
cancer_df.tail()
```

```python
allsites_10yr_avg_pc = []
lungBronchus_10yr_avg_pc = []


for x in range(len(cancer_df)):
    allsites_pc = round((cancer_df['2011_AllSites_perCapita'][x]
                   + cancer_df['2012_AllSites_perCapita'][x]
                   + cancer_df['2013_AllSites_perCapita'][x]
                   + cancer_df['2014_AllSites_perCapita'][x]
                   + cancer_df['2015_AllSites_perCapita'][x]
                   + cancer_df['2016_AllSites_perCapita'][x]
                   + cancer_df['2017_AllSites_perCapita'][x]
                   + cancer_df['2018_AllSites_perCapita'][x]
                   + cancer_df['2019_AllSites_perCapita'][x]
                   + cancer_df['2020_AllSites_perCapita'][x]) / 10,6)

    allsites_10yr_avg_pc.append(allsites_pc)

    lung_bronchus_pc = round((cancer_df['2011_LungBronchus_perCapita'][x]
                        + cancer_df['2012_LungBronchus_perCapita'][x]
                        + cancer_df['2013_LungBronchus_perCapita'][x]
                        + cancer_df['2014_LungBronchus_perCapita'][x]
                        + cancer_df['2015_LungBronchus_perCapita'][x]
                        + cancer_df['2016_LungBronchus_perCapita'][x]
                        + cancer_df['2017_LungBronchus_perCapita'][x]
                        + cancer_df['2018_LungBronchus_perCapita'][x]
                        + cancer_df['2019_LungBronchus_perCapita'][x]
                        + cancer_df['2020_LungBronchus_perCapita'][x]) / 10,6)

    lungBronchus_10yr_avg_pc.append(lung_bronchus_pc)
```

```python
cancer_df['2011-2020 allsites/capita AVG'] = allsites_10yr_avg_pc
cancer_df['2011-2020 lungBronchus/capita AVG'] = lungBronchus_10yr_avg_pc
cancer_df.head()
```

```python
all_classifier =[]
lung_bronchus_classifier = []
for x in range(len(cancer_df)):
    if cancer_df['2021_AllSites_perCapita'][x] > cancer_df['2011-2020 allsites/capita AVG'][x]:
        all_classification = 1
        all_classifier.append(all_classification)
    elif cancer_df['2021_AllSites_perCapita'][x] < cancer_df['2011-2020 allsites/capita AVG'][x]:
        all_classification = 0
        all_classifier.append(all_classification)
    else:
        all_classification = 'x'
        all_classifier.append(all_classification)

for x in range(len(cancer_df)):
    if cancer_df['2021_LungBronchus_perCapita'][x] > cancer_df['2011-2020 lungBronchus/capita AVG'][x]:
        lb_classification = 1
        lung_bronchus_classifier.append(lb_classification)
    elif cancer_df['2021_LungBronchus_perCapita'][x] < cancer_df['2011-2020 lungBronchus/capita AVG'][x]:
        lb_classification = 0
        lung_bronchus_classifier.append(lb_classification)
    else:
        lb_classification = 'x'
        lung_bronchus_classifier.append(lb_classification)
```
