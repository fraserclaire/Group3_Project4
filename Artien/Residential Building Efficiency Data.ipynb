```python
# According to the documentation, any cell with a 'Q' signifies that data was witheld because the RSE (relative standard error)
# was greater than 50% OR fewer than 10 households were found in the reporting sample.
# N signifies that no households were in the reporting sample

# Either way, a Q or N means the data is statistically insignificant or an outlier.

# Loop through the data set and replace any instances of 'Q' or 'N' with a 0
for i in range(len(residential_final_df)):
    for j in range(len(residential_final_df.columns)):
        if residential_final_df.iloc[i, j] == 'Q' or residential_final_df.iloc[i, j] == 'N':
            residential_final_df.iloc[i, j] = 0

residential_final_df.tail()
```
