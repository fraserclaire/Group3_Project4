```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

energy_cancer = pd.read_csv("energy_cancer_statePerYearwPop.csv")
energy_cancer.head()

# Select only the columns with _AllSites_perCapita
cancer_rate_columns = [col for col in energy_cancer.columns if '_AllSites_perCapita' in col]

# Select only the columns related to cancer rates and the 'State Abbreviation'
all_sites = energy_cancer[['State Abbreviation'] + cancer_rate_columns]

# Set the 'State Abbreviation' as the index
all_sites.set_index('State Abbreviation', inplace=True)

# Exclude 'US' row
all_sites = all_sites[all_sites.index != 'US']
all_sites = all_sites[all_sites.index != 'DC']

all_sites_percentage = all_sites * 100
all_sites_percentage

# List of years
years = [str(year) for year in range(2010, 2021)]

# Set seaborn style
sns.set_style("whitegrid")

# Plotting
plt.figure(figsize=(9, 6))
sns.boxplot(data=all_sites_percentage, palette="YlOrRd", linewidth=2, showfliers=True)
plt.title('Rates of All Cancer Sites - United States', fontsize=15, fontweight='bold')
plt.ylabel('Cancer Rate (%)', fontsize=14)
plt.xticks(range(11), years, fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()

# Save the plot as PNG
plt.savefig('allcancer_rates.png')

plt.show()
```
