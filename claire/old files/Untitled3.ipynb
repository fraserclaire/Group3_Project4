```python
from pyspark import SparkFiles
url ="https://raw.githubusercontent.com/fraserclaire/Group3_Project4/main/Resources/energy_cancer_statePerYear.csv"

spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("energy_cancer_statePerYear.csv"), sep=",", header=True, ignoreLeadingWhiteSpace=True) #Observe the need to use ignoreLeadingWhiteSpace=True, otherwise a leading whitespace will appear in the column names

# Show DataFrame
df.show()
```
