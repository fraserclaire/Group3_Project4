```python
#Import Dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
from pathlib import Path
```

```python
# The path to our CSV file
cancer_data = Path("cancerdata.csv")

# Read our cancer data into pandas
cancer_df = pd.read_csv(cancer_data)
cancer_df.head()
```

```python
#Po
```

```python
cancer_data_df = cancer_df.dropna(how="any")
cancer_data_df.head()
```

```python

cancer_bar = cancer_data_df.plot(kind="bar", grid = True, figsize =(20,8), xlabel = 'State', ylabel = 'Average Annual Count')
cancer_bar.set_xticklabels(cancer_data_df['State'], rotation=90)



plt.show()
```
